###### Load the dataset
The transcript was cleaned manually using Microsoft Excel.

```python
df = pd.read_csv('cleaned_transcript.csv')
df.head()
```

The NLTK Python library, leveraging the VADER lexicon, was used to compute the sentiment for each question and response.

_Hutto, Clayton J. and Gilbert, Eric. "VADER: A Parsimonious Rule-Based Model for Sentiment Analysis of Social Media Text.." Paper presented at the meeting of the ICWSM, 2014._

```python
sid = SentimentIntensityAnalyzer()
sentiment = sid.polarity_scores(df.Statement[0])
sentiment_score = sentiment['compound']

# Sample sentiment. Compound represents overall response sentiment 
# on a scale of -1.0 (extremely negative) to +1.0 (extremely positive)
print(df.Statement[0],sentiment)
```

###### Process entire transcript

```python
def process_statement(speaker, statement):
    # Calculate sentiment based on full string
    sentiment = sid.polarity_scores(statement)
    sentiment_score = sentiment['compound']
    
    # process data
    lower = statement.lower()
    words = lower.split()
    reformed = [appos[word] if word in appos else word for word in words]
    reformed = " ".join(reformed)
    
    stop_words = stopwords.words('english')
    
    # tokenize data and perform lemmatization
    tokens = word_tokenize(reformed)
    nostops = [i for i in tokens if i not in stop_words]
    words = [word for word in nostops if word.isalpha()]
    # lemm_full = [x for x in lemmatize(lower)]
    #lemm = [x.decode('utf-8').split('/') for x in lemmatize(str(lower))]
    #lem_words = [x[0] for x in lemm]
    #lem_tags = [x[1] for x in lemm]
    
    return [speaker, statement, sentiment_score, words]
```

```python
processedTranscript = pd.DataFrame(columns=['speaker','statement', 'sentiment','words'])

for index,row in df.iterrows():
    speaker = row['Speaker']
    statement = row['Statement']
    processedTranscript.loc[len(processedTranscript)] = process_statement(speaker, statement)
```

For this exercise, we will only be worried about looking at the candidates from the first debate. As such, we can remove the debate moderators.

```python
candidates = processedTranscript[~processedTranscript.speaker.isin(['HOLT', 'GUTHRIE', 'MADDOW', 'TODD', 'DIAZ-BALART','UNKNOWN'])]
candidates.head()
```

###### Responses by candidate
According to the transcript, Beto O'Rourke had the most separate moments of speaking, while Jay Inslee spoke the least, only appearing in the transcript on eight separate occassions.

```python
# Statements by Candidate
candidates[['speaker', 'statement']].groupby(['speaker']).agg(['count'])
```

When taking an average sentiment of all statements made by each candidate, Jay Inslee was, on average, the most positive (although his above mentioned small sample size no doubt skewed his average higher).



Tulsi Gabbard had the lowest average sentiment score at -11.9%. While having over twice as many speaking opportunities as Inslee, she was still well below the average number of opportunities compared to the rest of the field.

The terms positive and negative are being used in context to the dataset, as the average for each candidate could be better described as "neutral."

```python
candidates[['speaker', 'sentiment']].groupby(['speaker']).agg(['mean'])
```

###### Sentiment Swings
Every candidate on stage at the first debate registered extemely negative and positive sentiment ratings. While Inslee had the smallest swing among the candidates, his low sample size again makes him an outlier.

```python
# lowest and highest sentiment by speaker
maxPos = candidates[['speaker', 'sentiment']].groupby(['speaker'], as_index=False).agg(['max'])
maxNeg = candidates[['speaker', 'sentiment']].groupby(['speaker'], as_index=False).agg(['min'])
pd.concat([maxNeg,maxPos],axis=1)
```

###### Candidate sentiment trends by statement over course of debate.
The line plots show the sentiment for each response given by a candidate over the course of the debate. The shift between negative and positive sentiments were apparant for each candidate.

The sentiments of Booker, DeBlasio, Klobuchar and Ryan trended relatively flat over the course of the debate, while Castro trended downward before leveling off.

Warren's highest moment in terms of sentiment came at the start of the night before moving below zero and trending upward through towards the end of the debate, While O'Rourke remained relatively flat during the first half of the debate, he trended upward steadily to end with some of his highest sentiment scores of the night (including the above max sentiment score).
